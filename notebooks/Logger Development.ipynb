```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
vals = [50, 60, 70]
labels=['carbs','protein','calories']
pos = arange(len(vals))

rects = plt.barh(pos, vals, align='center')
yticks(pos, labels)
xlabel('Percent Fulfilled')
plt.axvline(x=100, linewidth=4, color='r')
plt.savefig('figure.png')
plt.axis([0, 120, -.5, len(vals)-.5])
xticks([20, 40, 60, 80, 100],[20, 40, 60, 80, 'Daily Req'])

for rect, label in zip(rects, labels):
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2, rect.get_y() + rect.get_height()/2 - .05, label, ha='center', va='bottom', color='w')
```

```python
from pylab import *
val = 3+10*rand(5)    # the bar lengths
pos = arange(5)+.5    # the bar centers on the y axis

figure(1)
barh(pos,val, align='center')
yticks(pos, ('Tom', 'Dick', 'Harry', 'Slim', 'Jim'))
xlabel('Performance')
title('How fast do you want to go today?')
grid(True)

figure(2)
barh(pos,val, xerr=rand(5), ecolor='r', align='center')
yticks(pos, ('Tom', 'Dick', 'Harry', 'Slim', 'Jim'))
xlabel('Performance')

show()
```
